```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

image.png

```python
# Data Pelatihan
X_train = np.array([[1.0, 1.0], [2.0, 2.0], [1.0, 3.0], [3.0, 3.0], [2.0, 1.0]])
y_train = np.array([0, 1, 0, 1, 0])
```

```python
# Titik Data Baru
X_new = np.array([[1.5, 2.0]])
```

image.png

```python

# Inisialisasi model KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Melatih model
knn.fit(X_train, y_train)
```

image.png

```python
# Prediksi menggunakan data baru
y_pred = knn.predict(X_new)

# Output hasil prediksi
print(f"Titik data baru {X_new} diklasifikasikan sebagai kelas {y_pred[0]}")
```
